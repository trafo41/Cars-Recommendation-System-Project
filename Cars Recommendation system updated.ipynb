```python
final_df = df_car_correlation.T
final_df.columns = df_cars.index
final_df.head()
```

```python
test_original = test.pivot(
    index = 'Cars name',
    columns = 'Users',
    values = 'Ratings'
)
```

```python
from numpy import *
total_non_nan = np.count_nonzero(~np.isnan(y2))
```

```python
import math

lista = []
for i in range(0,125):
    templist = [b / m for b,m in zip(list(y2.iloc[i].values),list(test_original.iloc[i].values))]
    cleanedList = [x for x in templist if (math.isnan(x) == False)]
    lista.append(sum(cleanedList))
#   len(lista)
```
