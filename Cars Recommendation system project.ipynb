# Part 2 : Item based collaborative filtering

```python
this_df = pd.read_csv('new_cars.csv')
this_df.head()
```

```python
df1 = pd.DataFrame(this_df.melt(id_vars = ["Users"]))
df1.columns = ['Users', 'Cars name', 'Ratings']
df1.head()
```

```python
df1.shape
```

```python
df_cars = df1.pivot(
    index = 'Cars name',
    columns = 'Users',
    values = 'Ratings'
).fillna(0)
```

```python
df_cars.head()
```

```python
from sklearn.metrics.pairwise import pairwise_distances

# Cars Similarity Matrix
car_correlation = 1 - pairwise_distances(df_cars, metric='cosine')
car_correlation[np.isnan(car_correlation)] = 0
print(car_correlation)
```

```python
car_correlation[car_correlation < 0] = 0
car_correlation
```

```python
car_correlation.shape
```

```python
df_car_correlation = pd.DataFrame(car_correlation)
df_car_correlation.columns = df_cars.index
df_car_correlation.head()
```

```python
final_df = df_car_correlation.T
final_df.columns = df_cars.index
final_df.head()
```

```python
test_original = test.pivot(
    index = 'Cars name',
    columns = 'Users',
    values = 'Ratings'
)
```

```python
from numpy import *
total_non_nan = np.count_nonzero(~np.isnan(y2))
```

```python
import math

lista = []
for i in range(0,125):
    templist = [b / m for b,m in zip(list(y2.iloc[i].values),list(test_original.iloc[i].values))]
    cleanedList = [x for x in templist if (math.isnan(x) == False)]
    lista.append(sum(cleanedList))
#   len(lista)
```

```python
accuracy = (sum(lista)/total_non_nan)*100
print("Accuracy of this model is : {} %".format(np.round(accuracy, 2)))
```
